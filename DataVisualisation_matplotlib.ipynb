## Histograms

What histogram plots is frequency of any column or feature w.r.t its range of values.

x-> range of values of feature

y-> frequency of that feature in the particular range

```
## ONe way of doing
print(iris.plot.hist())
## Another way of doing
print(iris.plot(kind='hist'))
```

# HIstogram manipulation.

stacked(True): then it will show the ovelapping information too.

stacked(False): it wont show the overlapping information.

bins: greater the bins more will be the informations

```
print(iris.plot.hist(stacked=True,bins=100,figsize=(15,10)))
print(iris.plot.hist(stacked=False,bins=100,figsize=(15,10)))
```

# df.diff()

this shows the difference of the column value to the next value in the same column.

```
iris_diff=iris['petal_length'].diff()
print(iris_diff.head(10))
iris_diff.diff().plot(kind='hist',stacked=True,bins=50,figsize=(15,10),title='iris dataset')
```

# Scatter plot

```
print(iris_df.plot(kind='scatter',x='sepal_length',y='petal_length'))

## This is a simple graph which is use to correlate two features 
iris.plot.scatter(x='sepal_width',y='petal_width')
```

Using third variable or you can say z axis in scatter plot.

```
print('\n--> In 1st plot c is taken as color variable:\n','-'*60)
print('--> In second plot c is taken as third variable:\n','-'*60)
print(iris_df.plot.scatter(x='sepal_length',y='petal_length',c='#a98d19' ))
print(iris_df.plot.scatter(x='sepal_length',y='petal_length',c='petal_width' ))
```

## Giving Label an size of the dots in the plot

```
iris_df.plot.scatter(x='sepal_length',y='petal_length',label='length',s=100)
## here label: labels the graph
## s: determines the size of the dots we want to keep.
```

# Pie plot

```
## this shows basically the share of data among all the plotting elements.
print(iris_df.iloc[0])
tot=iris_df.iloc[0].sum()
print('\nSum of all the elements\n','-'*50)
print(tot)
print('-'*50)   
print('Showing percentage share of each element\n','-'*50)
for i in iris_df.iloc[0]:
   print((i/(tot/100)).round(2),end='  ')
print('\n','-'*50)
print('Plotting in pie chart\n','-'*50)   
iris_df.iloc[0].plot.pie(figsize=(10,10))
```
