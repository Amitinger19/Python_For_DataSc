# Copy operations

--> Shallow copy

--> Deep copy


```
## Shallow copy
import numpy as np

x=np.array([1,2,3])
y=x ## this is shallow copy, both x and y are in the same memory so changing one with change the other.
y[0]=100
print(x) ## out:array([100,   2,   3]) so changing any value in y will change the value in x.



# Deep copy
# here z is stored in different memmory altogether so z and x are different can be accessed separately
z=np.copy(x)
z[0]=200
print(x)
print('\n',z)


## chacking each by asarray()

print(np.asarray(z) is x) ## False
print(np.asarray(y) is x) ## True
## here is operator comparing the address 
print(id(x))  #140249298487456
print(id(y))  #140249298487456
print(id(z))  #140249298488176
 
```

# np.diag(matrix,k=value)


-->if  k=+ve 

   means we will get the diagonal element of matrix removing k rows from bottom of the matrix and removing k columns from left of the matrix.

--> if k= -ve
   
   removes k rows from top of the matrix and removes k columns from right of the matrix.

In simple words k is the k th diagonal of the matrix, where k=0  for main diagonal.      

```
print(np.diag(check_mat,k=1))

## doubt: what if the matrix is not square?
print('\n'*3)
diag_ele=np.diag(check_mat,k=-2)
print(diag_ele)
```

# np.diag(np.diag(x))

Applying np.diag() twice, it will return a matrix with diagonal element given by np.diag(x). Where x is the matrix passed.

```
diag_element=np.diag(check_mat)
print(diag_element)
print('\n'*3)
double_diag_mat=np.diag(np.diag(check_mat))
print(double_diag_mat)
```

matrix.max(), matrix.min(), matrix.argmax()

```
 ## max(): this return the actual max value
x=np.random.randint(1,100,(4,5))
print('\nMatrix  x\n','-'*50)
print(x)
print('\nMax value in whole x\n','-'*50)
print(x.max())
print('\nMax value in x by row wise \n','-'*50)
print(x.max(axis=0))
print('\nMax value in x column wise\n','-'*50)
print(x.max(axis=1))

## argmax() this returns the location of maximium value
print('\nlocation Max value in x column wise\n','-'*50)
print(x.argmax(axis=1))
print('\nlocation Max value in x row wise\n','-'*50)
print(x.argmax(axis=0))
```
