## pandas.Categorical

What it does is  it shows the category of data in a categorical col.




```
new_class=pd.Categorical(df['Pclass'])
new_class
"""
[3, 1, 3, 1, 3, ..., 2, 1, 3, 1, 3]
Length: 891
Categories (3, int64): [1, 2, 3]

"""
# Categories (3, int64): [1, 2, 3] this means that 'Pclass' has 3 categories of data of int type

#If I want to change the calue of the categories to some other value
# Then that is very much doable by the help of following commanng
new_class=new_class.rename_categories(['a','b','c'])
# This will rename the [1,2,3] categories to ['a','b','c'] order wise
new_class
df['Pclass']=new_class  #This replaces the old value with new value in dataframe  
df[4:9]
```

```
## Accessing unique elements in any column.

df['Pclass'].unique()
```
